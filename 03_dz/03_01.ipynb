```python
%matplotlib inline 
import numpy as np 
import matplotlib.pyplot as plt 
k1 = a1 = b1 = 1 
k2 = a2 = b2 = 2 
k3 = a3 = b3 = 3 
x = np.linspace(-np.pi, np.pi, 201) 
plt.plot(x, k1*np.cos(x-a1)+b1) 
plt.plot(x, k2*np.cos(x-a2)+b2) 
plt.plot(x, k3*np.cos(x-a3)+b3) 
plt.xlabel("x") 
plt.ylabel("y") 
plt.grid(True) 
```
