```python
#2. Напишите код, который будет рисовать график окружности в полярных координатах. 
```

```python
%matplotlib inline 
import numpy as np 
import matplotlib.pyplot as plt 
import math 
R = 5 
t = np.linspace(-5, 5, 100) 
x = R*np.cos(t) 
y = R*np.sin(t) 
plt.plot(x,y) 
plt.xlabel("x") 
plt.ylabel("y") 
```
