```python
%matplotlib inline 
import numpy as np 
import matplotlib.pyplot as plt 
import math 
```

```python
R = 5 
t = np.linspace(-5, 5, 100) 
x = R*np.cos(t) 
y = R*np.sin(t) 
plt.plot(x,y) 
plt.xlabel("x") 
plt.ylabel("y") 
```

```python
a = 5 
b = 2.5 
t = np.linspace(-5, 5, 100) 
x = a*np.cos(t) 
y = b*np.sin(t) 
plt.plot(x,y) 
plt.xlabel("x") 
plt.ylabel("y") 
```
