```python
%matplotlib inline 
import numpy as np 
import matplotlib.pyplot as plt 
from pylab import * 
from mpl_toolkits.mplot3d import Axes3D 
```

```python
fig = figure() 
ax = Axes3D(fig)
X = np.arange(-5, 5, 0.5) 
Y = np.arange(-5, 5, 0.5) 
X, Y = np.meshgrid(X, Y) 
Z = 2*X + Y 
Z2 = (-1/2)*X + Y 
ax.plot_surface(X, Y, Z) 
ax.plot_surface(X, Y, Z2) 
```

```python
#2. Нарисуйте трехмерный график двух любых поверхностей второго порядка. 
```

```python
fig = figure() 
ax = Axes3D(fig)
X = np.arange(-5, 5, 0.5) 
Y = np.arange(-5, 5, 0.5) 
X, Y = np.meshgrid(X, Y) 
Z = X**2 + Y**2 
Z2 = -(X**2 + Y**2) 
ax.plot_surface(X, Y, Z) 
ax.plot_surface(X, Y, Z2) 
```
